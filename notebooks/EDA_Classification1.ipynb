## 4. Univariate Analysis

### 4.1. Histograms of each Variable

```python
df_raw_numeric = df_raw.select_dtypes(include="number")

for column in df_raw_numeric:
    plt.title(column)
    plt.hist(df_raw[column])
    plt.tight_layout()
    plt.show()
```

### 4.2. Outlier detection via boxplot

```python
for column in df_raw_numeric:
    plt.title(column)
    plt.boxplot(df_raw[column])
    plt.tight_layout()
    plt.show()
```
